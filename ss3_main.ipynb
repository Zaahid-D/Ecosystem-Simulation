# Explanation of the Key Parameters:

## Average Birth Rate: 0.3

Herbivorous species such as zebras and buffalo exhibit relatively high reproductive rates. For instance, zebras can reproduce annually, which significantly contributes to their population growth. In contrast, carnivorous species like lions, as well as large mammals like elephants, have lower reproductive rates. Elephants, for example, give birth only once every few years, which greatly influences their population dynamics. The chosen birth rate of 0.3 represents a balanced average that reflects the higher birth rates typical of herbivores and the lower birth rates of larger mammals and carnivores. This average rate is intended to provide a realistic simulation of population growth across different species.

## Average Mortality Rate: 0.15

In the wild, many young animals experience high mortality rates due to predation and environmental factors, but adult mortality is generally lower as they become more resilient and less vulnerable to predators. Different species exhibit varying mortality risks. For example, elephants, which have few natural predators, generally experience lower mortality rates. On the other hand, species like zebras face significant predation pressure, especially in their juvenile stages, leading to higher mortality rates. The selected mortality rate of 0.15 is a reasonable average that balances the high juvenile mortality and the lower adult mortality rates across different species in the savanna ecosystem.

## Average Migration Rate: 0.03

Many savanna species, such as zebras and buffalo, undertake seasonal migrations in search of water and food, which are crucial for their survival in the varying climatic conditions of the savanna. However, some species, like lions, are more territorial and exhibit less migration. Their movement is generally confined to their established territories. The chosen migration rate of 0.03 reflects a moderate level of migration that accounts for the significant seasonal movements of some species without excessively reducing the population due to migration. This rate aims to provide a realistic depiction of the seasonal dynamics and resource-driven movements within the ecosystem.

## Time Steps: 15 Years

The simulation will be conducted over a period of 15 years. This timeframe allows us to observe and analyze long-term trends and dynamics within the savanna ecosystem. By simulating population changes, birth rates, mortality rates, and migration patterns over these 15 years, we can gain valuable insights into the behavior and interactions of different species in their natural habitat.

## Initial Population Size: 1 000

An initial population of 1000 individuals provides a robust starting point for simulating the dynamics of the savanna ecosystem. This population size is large enough to accurately capture the interactions and variations among different species. With 1000 individuals, we can observe meaningful trends in birth rates, mortality rates, and migration patterns over the 15-year simulation period.
